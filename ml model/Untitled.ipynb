```python
df.rename(columns={'Productivity': 'productivity'}, inplace=True)
```

```python
import matplotlib.pyplot as plt
import random

# Define ranges for data (assuming these values are not loaded from data)
it_min, it_max = 1, 5
productivity_min, productivity_max = 30, 80
noise_range = (-0.5, 0.5)  # Range for random noise

# Function to generate productivity based on IT with noise
def get_productivity(it):
  base_productivity = (it - it_min) * (productivity_max - productivity_min) / (it_max - it_min)
  noise = random.uniform(*noise_range)
  return base_productivity + noise

# Generate data (assuming you don't have existing data)
data = []
for _ in range(1000):
  it = random.randint(it_min, it_max)
  productivity = get_productivity(it)
  data.append((it, productivity))

# Load data from CSV (modify as needed)
# with open("it_productivity.csv", "r") as f:
#   reader = csv.reader(f)
#   next(reader)  # Skip header row
#   data = list(reader)

# Assuming your data is loaded into a list called 'data'

# Extract data for plotting (assuming columns at indices 0 and 1)
it_values = [row[0] for row in data]
productivity_values = [row[1] for row in data]

# Print data for verification (optional)
# print(f"IT values: {it_values[:5]}")
# print(f"Productivity values: {productivity_values[:5]}")

# Configure plot
plt.figure(figsize=(12, 8))  # Increased figure size for better readability
plt.scatter(it_values, productivity_values, alpha=0.8, color='royalblue', marker='o', edgecolors='black')  # Adjusted aesthetics
plt.title("Information Technology vs. Productivity", fontsize=18, fontweight='bold')  # Enhanced title
plt.xlabel("Information Technology (IT)", fontsize=16)  # Clearer axis labels
plt.ylabel("Productivity", fontsize=16)

# Grid and Ticks
plt.grid(True, linestyle='--', linewidth=0.7, color='lightgray', alpha=0.7)  # Finer grid lines
plt.xticks(range(it_min, it_max + 1))  # Set x-axis ticks for all IT values
plt.yticks(range(int(productivity_min), int(productivity_max) + 1, 5))  # Set y-axis ticks with spacing

# Customize plot aesthetics (optional)
plt.xlim(it_min - 0.2, it_max + 0.2)  # Adjusted x-axis limits for better margins
plt.ylim(productivity_min - 5, productivity_max + 5)  # Adjusted y-axis limits for better margins

plt.show()
```
